# Classe com Maior Número de Mortes
Analisar a coluna 'Pclass' e 'Survived' para determinar a classe com maior número de mortes.

```python
# Analisar a coluna 'Pclass' e 'Survived' para determinar a classe com maior número de mortes
mortes_por_classe = df[df['Survived'] == 0]['Pclass'].value_counts()
classe_maior_mortes = mortes_por_classe.idxmax()
print(classe_maior_mortes)
```

# Idade Média das Pessoas que Morreram
Calcular a idade média das pessoas que morreram utilizando df[df['Survived'] == 0]['Age'].mean().

```python
# Calcular a idade média das pessoas que morreram
idade_media_mortos = df[df['Survived'] == 0]['Age'].mean()
print(idade_media_mortos)
```

# Verificar Dados Duplicados e Nomes Repetidos
Verificar dados duplicados com df.duplicated().sum() e nomes repetidos com df['Name'].duplicated().sum().

```python
# Verificar Dados Duplicados e Nomes Repetidos

# Verificar dados duplicados
dados_duplicados = df.duplicated().sum()
dados_duplicados

# Verificar nomes repetidos
nomes_repetidos = df['Name'].duplicated().sum()
print(nomes_repetidos)
```

# Verificar e Tratar Valores Nulos
Verificar valores nulos com df.isnull().sum() e sugerir tratamentos como preenchimento ou remoção.

```python
# Verificar e Tratar Valores Nulos

# Verificar valores nulos em cada coluna
valores_nulos = df.isnull().sum()
valores_nulos

# Sugerir tratamentos para valores nulos
# Preencher valores nulos na coluna 'Age' com a média das idades
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Preencher valores nulos na coluna 'Embarked' com o valor mais frequente
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Preencher valores nulos na coluna 'Cabin' com 'Unknown'
df['Cabin'].fillna('Unknown', inplace=True)

# Verificar novamente valores nulos após o tratamento
valores_nulos_tratados = df.isnull().sum()
valores_nulos_tratados
```

# Transformar Variáveis Categóricas em Numéricas
Utilizar pd.get_dummies() para transformar variáveis categóricas de linguagem e gênero em numéricas.

```python
# Transformar Variáveis Categóricas em Numéricas

# Utilizar pd.get_dummies() para transformar variáveis categóricas de linguagem e gênero em numéricas
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)

# Exibir as primeiras linhas do dataset para verificar as transformações
df.head()
```

# Normalizar Variáveis Numéricas
Normalizar variáveis numéricas utilizando MinMaxScaler do sklearn.

```python
from sklearn.preprocessing import MinMaxScaler

# Selecionar apenas as colunas numéricas
colunas_numericas = df.select_dtypes(include=[np.number]).columns

# Inicializar o MinMaxScaler
scaler = MinMaxScaler()

# Aplicar o MinMaxScaler nas colunas numéricas
df[colunas_numericas] = scaler.fit_transform(df[colunas_numericas])

# Exibir as primeiras linhas do dataset para verificar a normalização
df.head()
```

# Gerar Gráficos e Insights
Gerar gráficos utilizando matplotlib e seaborn para obter insights dos dados.

```python
import seaborn as sns

# Configurar o estilo dos gráficos
sns.set(style="whitegrid")

# Gráfico de barras para mostrar o número de sobreviventes e não sobreviventes
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', data=df)
plt.title('Número de Sobreviventes e Não Sobreviventes')
plt.xlabel('Sobreviveu')
plt.ylabel('Contagem')
plt.show()

# Gráfico de barras para mostrar o número de mortes por classe
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Número de Mortes por Classe')
plt.xlabel('Classe')
plt.ylabel('Contagem')
plt.legend(title='Sobreviveu', loc='upper right')
plt.show()

# Gráfico de distribuição para mostrar a idade das pessoas que morreram
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Survived'] == 0]['Age'], bins=30, kde=True)
plt.title('Distribuição de Idade das Pessoas que Morreram')
plt.xlabel('Idade')
plt.ylabel('Densidade')
plt.show()

# Gráfico de barras para mostrar a distribuição de gênero entre sobreviventes e não sobreviventes
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex_male', hue='Survived', data=df)
plt.title('Distribuição de Gênero entre Sobreviventes e Não Sobreviventes')
plt.xlabel('Gênero (0 = Feminino, 1 = Masculino)')
plt.ylabel('Contagem')
plt.legend(title='Sobreviveu', loc='upper right')
plt.show()

# Gráfico de barras para mostrar a distribuição de embarque entre sobreviventes e não sobreviventes
plt.figure(figsize=(10, 6))
sns.countplot(x='Embarked_Q', hue='Survived', data=df)
plt.title('Distribuição de Embarque entre Sobreviventes e Não Sobreviventes')
plt.xlabel('Embarque (0 = Cherbourg/Southampton, 1 = Queenstown)')
plt.ylabel('Contagem')
plt.legend(title='Sobreviveu', loc='upper right')
plt.show()

# Gráfico de dispersão para mostrar a relação entre idade e tarifa
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Relação entre Idade e Tarifa')
plt.xlabel('Idade')
plt.ylabel('Tarifa')
plt.legend(title='Sobreviveu', loc='upper right')
plt.show()
```
